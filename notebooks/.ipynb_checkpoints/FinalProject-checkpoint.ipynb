# Are Greenland Glacial Earthquakes Remotely Triggered by Large Earthquakes?   

### RCES Final Project :: Theresa Sawi :: Tues December 4, 2018  


Large calving events at marine-terminating glaciers produce vibrations in the solid earth known as "glacial earthquakes", detectable at large distances and cataloged by the Harvard's Global CMT (from 1993-2013). I want to investigate if large earthquakes can trigger the occurrence of glacial earthquakes at Greenland. 

I will use the Global CMT catalog of glacial earthquakes as well as the USGS global earthquake catalog (both from 1993-2013) to investigate any possible temporal relationships between the two groups.

A similar phenomenon has been observed in Antarctica, wherein "icequakes" (seismicity caused by shallow tensile ice fracture) appeared to have been triggered by the 2010 M8.8 Maule earthquake in Chile (Peng et al., 2014).

```python
from matplotlib import pyplot as plt
from matplotlib import cm
import matplotlib
import pandas as pd
import cartopy.crs as ccrs
import cartopy
import numpy as np
from datetime import datetime
from haversine import haversine



# #see all output, not just last command
# from IPython.core.interactiveshell import InteractiveShell
# InteractiveShell.ast_node_interactivity = "all"
```

# Data Wrangling


From USGS (https://earthquake.usgs.gov) and Harvard Global CMT (https://www.globalcmt.org/CMTsearch.html)

```python
########################
###LOAD DATA SETS    
########################

# Global catalogs 
# M8_df = pd.read_csv('../data/eqs/queryM8.csv')

#Greenland tect eq catalogs
# [50.736, 85.051] Latitude
# [-100.547, 21.797] Longitude
M8_df = pd.read_csv('../data/eqs/queryGL_M5.csv')# 
# M8_df = pd.read_csv('../data/eqs/queryGL_M4.csv')# 

#Greenland Icequakes
GL_df = pd.read_csv('../data/eqs/CMT_glacialEQ.csv')




########################
###SET PARAMETER HERE###
########################
minMag = 5.5 #Min mag of tectonic event






########################
###FORMAT TIME ###
########################

#reformat time column for Harvard CMT glacial eq catalog to be like USGS:'2013-05-24T05:44:48.980Z'
datetime_temp = []
for i in range(len(GL_df)):
    year = str(GL_df['year'][i])
    month = str(GL_df['month'][i]).zfill(2)
    day = str(GL_df['day'][i]).zfill(2)
    hour = str(GL_df['hour'][i]).zfill(2)
    minute = str(GL_df['minute'][i]).zfill(2)
    second = str(GL_df['second'][i]).zfill(2)
    microsecond = str('.' + '000')
    datetime_temp.append(str(year) + '-' + str(month)  + '-' + str(day) + 'T'\
                     + str(hour) + ':' + str(minute) + ':' +str(second) + microsecond +'Z')

    
    
#convert to datetime type, sort
GL_df['time'] = datetime_temp
GL_df['time'] = pd.to_datetime(GL_df['time'])
GL_df = GL_df.sort_values(by='time', ascending=False)

M8_df['time'] =  pd.to_datetime(M8_df['time'])
M8_df = pd.DataFrame(M8_df.sort_values(by='time', ascending=False))




########################
###preview datasets
########################

# print('Tectonic earthquakes, 1993-2013')
# print(M8_df.head(1))
# M8_df.tail(1)


# print('Glacial earthquakes, 1993-2013')
# print(GL_df.head(1))
# GL_df.tail(1)




M8_df = M8_df.loc[M8_df['mag'] >= minMag]

startyear = GL_df['year'].min()
endyear = GL_df['year'].max()
nYears = endyear - startyear
yearRange = str(startyear) + '-' + str(endyear) #handy for plotting
```

```python
print(str(len(GL_df)) + " glacial earthquakes")
print(str(len(M8_df)) + " tectonic earthquakes")
```

# Analysis:

### Find seismic events w/in #|dayRange|# days of eachother

```python
dayRange = 15

#make new Dataframe of IQs (icequakes) occurring within dateRange of tectonic eqs
dt = []     #time difference between iq and eq
iqt = []    #time iq
iqlat = []  #lat iq
iqlon = []  #lon iq
eqt = []    #time eq
eqlat = []  #lat eq
eqlon = []  #lon eq
eqmag = []  #eq mag
hav = []    #great circle distance btwn iq and eq


# # test seaonal sensitivity
# winter_months = [12,1,2,3]
# summer_months = [6,7,8]


for i in range(len(M8_df)):

    eqtime = M8_df['time'].iloc[i]    
    
    
    for j in range(len(GL_df)):
        
        iqtime = GL_df['time'].iloc[j]
        
        
        time_diff = iqtime - eqtime
        
        
        
        if (time_diff.total_seconds()/86400) <= dayRange and (time_diff.total_seconds()/86400) >= -dayRange:
            
            iclatt = GL_df['lat'].iloc[j]
            iclonn = GL_df['lon'].iloc[j]
            
            eqlatt = M8_df['latitude'].iloc[i]
            eqlonn = M8_df['longitude'].iloc[i]
            
            
            dt.append(time_diff.total_seconds()/86400)            #time difference, seconds
            hav.append(haversine((iclatt, iclonn),(eqlatt, eqlonn))) #haversine distance
            
            
            iqt.append(GL_df['time'].iloc[j])    #time iq
            iqlat.append(iclatt)  #lat iq
            iqlon.append(iclonn)  #lon iq
                
            eqt.append(M8_df['time'].iloc[i])    #time eq
            eqlat.append(eqlatt)                 #lat eq
            eqlon.append(eqlonn)                 #lon eq
            eqmag.append(M8_df['mag'].iloc[i])   #eq mag
   
        
          
    
len(dt)
```

```python
# manipulate columns into datafram
d = [dt,hav,eqt,eqlat,eqlon,eqmag,iqt,iqlat,iqlon]  #lon eq
dd = np.array(d) #np array must go into dataframe
ddT = dd.T # transpose to get the right shape
trig_table = pd.DataFrame(ddT, columns=['dt','hav','eqt','eqlat','eqlon','eqmag','iqt','iqlat','iqlon'])
trig_table_orig = trig_table

# len(trig_table_orig)
trig_table_orig.head()
```

##### Remove duplicate glacial eqs? So that a calving event is only triggered once.
This has implications for aftershocks, however, since systematically removing tectonic eq that are more likely to happen after glacial eq. 

```python
trig_table = trig_table.drop_duplicates('iqt')
# DataFrame.drop_duplicates(subset=None, keep=’first’, inplace=False)
```

## Effect of removing duplicated glacial eqs

Removing aftershocks may systematically alter the distribution of tectonic earthquakes with respect to glacial ones, so care must be taken when working with lower magnitudes.  

In this case, it seems that after deleting duplicated glacial events (that is, ones that were apparently "triggered" by more than one tectonic earthquake), more tectonic earthquakes are removed relatively after the glacial earthquake than before.   

Either way, there is no obvious triggering-relationship between tectonic and glacial earthquakes.  

We will investigate occurrences of this phenomenon in another geologc setting...
