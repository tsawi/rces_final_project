## Are Greenland Glacial Earthquakes Remotely Triggered by Large Earthquakes?   

### RCES Final Project :: Theresa Sawi :: Tues December 4, 2018  


Large calving events at marine-terminating glaciers produce vibrations in the solid earth known as "glacial earthquakes", detectable at large distances and cataloged by the Harvard's Global CMT (from 1993-2013). I want to investigate if large earthquakes can trigger the occurrence of glacial earthquakes at Greenland. 

I will use the Global CMT catalog of glacial earthquakes as well as the USGS global earthquake catalog (both from 1993-2013) to investigate any possible temporal relationships between the two groups.

A similar phenomenon has been observed in Antarctica, wherein "icequakes" (seismicity caused by shallow tensile ice fracture) appeared to have been induced by the 2010 M8.8 Maule earthquake in Chile (Peng et al., 2014).

In addition to vibrations passing through the solid earth, stresses could also be transmitted by ocean waves. To investigate this possibility, we will calculate the expected arrival of a wave using  (wavespeed between .5 and 50 m/s), and will note if any calving events occur in that window.

### Loop through events

```python
dayRange = 15

#make new DF of IQs triggered within dateRange
dt = []     #time difference between iq and eq
iqt = []    #time iq
iqlat = []  #lat iq
iqlon = []  #lon iq
eqt = []    #time eq
eqlat = []  #lat eq
eqlon = []  #lon eq
eqmag = []  #eq mag
hav = []    #great circle distance btwn iq and eq

count=0


# # test seaonal sensitivity
# winter_months = [12,1,2,3]
# summer_months = [6,7,8]


for i in range(len(M8_df)):

    eqtime = M8_df['time'].iloc[i]    
    
    
    for j in range(len(GL_df)):
        
        iqtime = GL_df['time'].iloc[j]
        
        
        time_diff = iqtime - eqtime
        
        
        
        if (time_diff.total_seconds()/86400) <= dayRange and (time_diff.total_seconds()/86400) >= -dayRange:
            
            iclatt = GL_df['lat'].iloc[j]
            iclonn = GL_df['lon'].iloc[j]
            
            eqlatt = M8_df['latitude'].iloc[i]
            eqlonn = M8_df['longitude'].iloc[i]
            
            
            dt.append(time_diff.total_seconds()/86400)            #time difference, seconds
            hav.append(haversine((iclatt, iclonn),(eqlatt, eqlonn))) #haversine distance
            
            
            iqt.append(GL_df['time'].iloc[j])    #time iq
            iqlat.append(iclatt)  #lat iq
            iqlon.append(iclonn)  #lon iq
                
            eqt.append(M8_df['time'].iloc[i])    #time eq
            eqlat.append(eqlatt)                 #lat eq
            eqlon.append(eqlonn)                 #lon eq
            eqmag.append(M8_df['mag'].iloc[i])   #eq mag

            
            
            
#         if (time_diff.total_seconds()/86400) <= 3 and (time_diff.total_seconds()/86400) > 0 :
#             print('Triggered glacial eq? Glacial eq = ' + str(j) + '; Major eq = ' + str(i))
#             print('Diff: ' + str(time_diff.days))
#             count += 1        
        
          
    
len(dt)
print(count)
```

```python
# manipulate columns into datafram
d = [dt,hav,eqt,eqlat,eqlon,eqmag,iqt,iqlat,iqlon]  #lon eq
dd = np.array(d) #np array must go into dataframe
ddT = dd.T # transpose to get the right shape
trig_table = pd.DataFrame(ddT, columns=['dt','hav','eqt','eqlat','eqlon','eqmag','iqt','iqlat','iqlon'])
trig_table_orig = trig_table

len(trig_table_orig)
trig_table_orig.head()
```

### Remove duplicate glacial eqs? So that a calving event is only triggered once.
This has implications for aftershocks, however, since systematically removing tectonic eq that are more likely to happen after glacial eq. 

```python
trig_table = trig_table.drop_duplicates('iqt')
# DataFrame.drop_duplicates(subset=None, keep=â€™firstâ€™, inplace=False)
```
