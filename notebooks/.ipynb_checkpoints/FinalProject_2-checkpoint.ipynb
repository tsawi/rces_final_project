```python
# Gorner Glacier : Get names of events to use as dates...

import glob



homedir = '/Users/temp/Documents/~CU_18Fa/' #RCES/rces-final-project/'    


pathIn = homedir + '~Research/gorner_2007_new/'


months = list(['May','Jun','Jul'])
days = []
event_times = []
index = 0
df = pd.DataFrame()

for mo in months:
    days.append(glob.glob(pathIn + mo + '/*'))
    for da in days:
        for d in da:
#             event_times.append(glob.glob(d + '/*'))
            event_times.append(glob.glob(d + '/filist'))
           
    
            df2 = pd.read_csv(event_times[index][0], header=None)
            df = df.append(df2)

            index += 1


numGorn = len(df)
numGorn
df.head(1)
```

```python
#reformat time column for Gorner Glacier eq catalog to be like USGS:'2013-05-24T05:44:48.980Z'

datetime_temp_gorner = []
n = 100

for i in range(len(df)):
    if np.mod(i,n) == 0:
        print(str(i) + ' of ' + str(len(df)))
        st = str(df.iloc[i]).split()[1].split('.')[0]


        year = str(st[0:2])
        month = str(st[2:4]).zfill(2)
        day = str(st[4:6]).zfill(2)
        hour = str(st[6:8]).zfill(2)
        minute = str(st[8:10]).zfill(2)
        second = str(st[10:12]).zfill(2)

        if int(second) >= 60:
            print('second >= 60 secs')
            pass

        us = str('.' + '000')


        datetime_temp_gorner.append(str(year) + '-' + str(month)  + '-' + str(day) + 'T'\
                         + str(hour) + ':' + str(minute) + ':' +str(second) + us +'Z')
```
