### Loop through events

```python
m8d1 = df_eq['time'].iloc[-1]
gld1 = df_ice['time'].iloc[-1]


m8d1
gld1


dt_diff = gld1 - m8d1

dt_diff

```

```python
#should make df of eqstarttime glaceqstarttime and dt

dt = []
gl = []
eq = []
dateRange = 10
count=0

winter_months = [12,1,2,3]

summer_months = [5,6,7,8]


for i in range(len(df_eq)):
# for i in range(1):

#     print(m8d)
    m8d1 = df_eq['time'].iloc[i]    
    
    
    for j in range(len(df_ice)):
#     for j in range(1):

        
        gld1 = df_ice['time'].iloc[j]
        
        
        dt_diff = gld1 - m8d1
        
        if np.abs(dt_diff.days) <= dateRange:
            dt.append(dt_diff.days)
            gl.append(gld1)
            eq.append(m8d1)
#             break
        
        if (dt_diff.total_seconds()/86400) <= 1 and (dt_diff.total_seconds()/86400) > 0 :
            print('Triggered glacial eq? Glacial eq = ' + str(j) + '; Major eq = ' + str(i))
            print('Diff: ' + str(dt_diff))
            count += 1        
        

        
        
        
        

# trig_table = pd.DataFrame( [dt,gl,eq],index=list(range(len(dt))), columns=['deltaT', 'glacT', 'eqT'])

        


    
    
len(dt)
print(count)
```
